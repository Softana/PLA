# Exercise 1 (20')

In the following code section:
1. Create a column vector with entries (3,2,1)
2. Create a row vector with entries (6,5)
3. Print the shapes of such vectors





```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots()

col_v = [3, 2, 1]
row_v = [6, 5]

ax.quiver(0, 0, col_v[0], col_v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Column Vector')

ax.quiver(0, 0, row_v[0], row_v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Row Vector')

ax.set_xlim([0, 10])
ax.set_ylim([0, 10])

ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.grid(True)
plt.legend()
plt.show()
```

# Exercise 2 (20')

Given vectors v and w defined below please compute the following operations:


1.   The addition of v and w
2.   The subtraction of v from w
3.   The multiplication of v by the scalar 9

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

v = np.array([[6],[10],[14]])
w = np.array([[4],[8],[12]])



```

# 1. The addition of v and w

```python
addition = v + w
print(addition)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, v[0,0], v[1,0], v[2,0], color='red', label='v', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, w[0,0], w[1,0], w[2,0], color='blue', label='w', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, addition[0,0], addition[1,0], addition[2,0], color='green', label='v+w', arrow_length_ratio=0.1)


ax.set_xlim([0, 20])
ax.set_ylim([0, 20])
ax.set_zlim([0, 30])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.grid(True)
ax.legend()
plt.show()
```

# 2. The subtraction of v from W

```python
subtraction = v - w
print(subtraction)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, v[0,0], v[1,0], v[2,0], color='red', label='v', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, w[0,0], w[1,0], w[2,0], color='blue', label='w', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, subtraction[0,0], subtraction[1,0], subtraction[2,0], color='green', label='v - w', arrow_length_ratio=0.2)


ax.set_xlim([0, 20])
ax.set_ylim([0, 20])
ax.set_zlim([0, 30])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.grid(True)
ax.legend()
plt.show()
```

# How to plot 2D vector

```python
import matplotlib.pyplot as plt
import numpy as np

v1 = [2, 5]
v2 = [6, 8]
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=2, color='red')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.xlim([0, 10])
plt.ylim([0, 10])

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
plt.show()
```

# How to plot a 3D vector


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

v1 = [1, 2, 3]
v2 = [1, 3, 2]

ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='red', arrow_length_ratio=0.2)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='blue', arrow_length_ratio=0.2)
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('3D Vector Plot')

plt.show()
```

# How to plot functions

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
plt.show()

```
